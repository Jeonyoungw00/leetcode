#####**[LeetCode 412](https://leetcode.com/problems/fizz-buzz/description/)**


```python
class Solution:
    def fizzBuzz(self, n: int) -> List[str]:
        aList=list()
        i = 1
        while i<=n:
            if i%3==0 and i%5==0:
                aList.append("FizzBuzz")
            elif i%5==0:
                aList.append("Buzz")
            elif i%3==0:
                aList.append("Fizz")
            else:
                aList.append(str(i))
            i+=1
        return aList
```
