#####**[LeetCode 392](https://leetcode.com/problems/is-subsequence/description/)**


```python
class Solution:
    def isSubsequence(self, s: str, t: str) -> bool:
        sp = tp = 0

        while sp < len(s) and tp < len(t):
            if s[sp] == t[tp]:
                sp += 1
            tp += 1

        return sp == len(s)
```
