#####**[LeetCode 171](https://leetcode.com/problems/excel-sheet-column-number/description/)**


```python
class Solution:
    def titleToNumber(self, columnTitle: str) -> int:
        ans, pos = 0, 0
        for letter in reversed(columnTitle):
            digit = ord(letter)-64
            ans += digit * 26**pos
            pos += 1

        return ans
```
