#####**[LeetCode 485](https://leetcode.com/problems/max-consecutive-ones/description/)**


```python
class Solution:
    def findMaxConsecutiveOnes(self, nums: List[int]) -> int:
        count = 0
        ans = 0
        for i in range(len(nums)):
            if nums[i] == 0:
                count = 0
            else:
                count += 1
                ans = max(ans, count)
        return ans
```
