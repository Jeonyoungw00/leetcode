#####**[LeetCode 217](https://leetcode.com/problems/contains-duplicate/description/)**


```python
class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:
        num_set = set()
        for i in nums:
            if i in num_set:
                return True
            else:
                num_set.add(i)
        return False
```
