#####**[LeetCode 883](https://leetcode.com/problems/projection-area-of-3d-shapes/description/)**


```python
class Solution:
    def projectionArea(self, grid: List[List[int]]) -> int:
        sum = len(grid)*len(grid[0])
        for i in grid:
            sum += max(i)
            sum -= i.count(0)

        for i in range(len(grid[0])):
            arr = []
            for j in grid:
                arr.append(j[i])
            sum += max(arr)

        return sum
```
